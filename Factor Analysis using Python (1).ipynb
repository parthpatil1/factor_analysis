```python
scaler = StandardScaler()
newdf = scaler.fit_transform(df)
```

```python
PCA=PCA()
PCA.fit(newdf)
```

```python
result = pd.DataFrame({
    "Eigen": PCA.explained_variance_,
    "Variance_ratio":PCA.explained_variance_ratio_ * 100,
    "CumulativeVariance": (PCA.explained_variance_ratio_ * 100).cumsum()
})
result.index = ['comp ' + str(i+1) for i in result.index]

result
```

```python
# Factor Analysis execution

fa = FactorAnalyzer(n_factors=5,rotation='varimax', method='principal')
fa.fit(df)

print(pd.DataFrame(fa.loadings_,index=df.columns))
```

```python
# Factor Scores

fscore = fa.transform(df)

fscore_data=pd.DataFrame(fscore)
fscore_data
```

```python
import statsmodels.api as sm
```

```python
# Add a constant term
factor_scores_with_constant = sm.add_constant(fscore_data)
y = df[['Rate your device on the basis of Overall performance']]
```

```python
# Run the regression
model = sm.OLS(y, factor_scores_with_constant)
results = model.fit()
```

```python
print(results.summary())
```
